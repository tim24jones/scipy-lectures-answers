```python
from scipy import misc
import numpy as np
%matplotlib inline
face=misc.face(gray=True)
```

```python
import pylab as plt
plt.imshow(face)
```

```python
plt.imshow(face, cmap=plt.cm.gray)
```

```python
crop_face = face[100:-100, 100:-100]
plt.imshow(face, cmap=plt.cm.gray)
```

```python
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]
y.shape, x.shape
```

```python
centerx, centery = (660, 300)
mask = ((y - centery)**2 + (x - centerx)**2) > 230**2
face[mask] = 0
plt.imshow(face, cmap=plt.cm.gray)
```

```python
centerx, centery = (660, 350)
mask = (2*(y - centery)**2 + (x - centerx)**2) > 230**2
face[mask] = 0
plt.imshow(face, cmap=plt.cm.gray)
```
